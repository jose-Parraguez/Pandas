```python
#libreria para hacer el analisis
import pandas as pd
import sqlite3


#Conexion con la base de datos de sqlite3
conexion = sqlite3.connect('Base de datos Examen.db')



# los datos se guardan en las varianles dAnimal y dzoo
Zoologico = pd.read_sql_query('SELECT * FROM Zoologico', conexion)
encargados = pd.read_sql_query('SELECT Animal,Encargado,Familia FROM Zoologico', conexion)
presupuestos = pd.read_sql_query('SELECT Presupuesto,Empresa,Continente,Pais FROM Zoologico', conexion) 

conexion.close()

#imprimo la varibable DF
Zoologico.head()
```

```python
#Filtra en los animales por pais_continente
resultados = Zoologico[Zoologico['Continente'] == 'America']
resultados.head()
```
